# Renaming the original column name to shorter name for easy handling

```python
# Check the original column names
print("Original column names:")
print(df.columns.tolist())

# Clean the column names (remove extra spaces and newlines)
df.columns = df.columns.str.strip().str.replace('\n', ' ')

# Define the column mapping
column_mapping = {
    'Age': 'Age',
    'Gender': 'Gender',
    'Highest educational Qualification': 'Education',
    'Experience in IT:' : 'Experience',
    'Designation': 'JobRole',
    'Are you aware of the interrelation between emotions and their impact on work performance?': 'EmoAwareness',
    'Do you believe that being aware of your emotions impacts your work?': 'EmoImpact',
    'Which emotions do you most commonly experience at work? (Select up to 3 options)': 'EmoStates',
    'Does your emotional state positively influence your work productivity?': 'EmoProductivity',
    'Are you able to remain calm and composed in the workplace despite environmental challenges?': 'CalmComposed',
    'Does your organization have policies or initiatives to enhance Emotional Intelligence or Emotional Self-Awareness?': 'OrgPolicies',
    'If yes, how effective are these policies in improving Emotional Intelligence?': 'PolicyEffectiveness',
    'How effectively can you manage stress during work challenges? (Rate on a scale of 1–5, where 1 = Poor and 5 = Excellent)': 'StressManagement',
    'How would you rate your tolerance and patience in handling workplace stressors?': 'TolerancePatience',
    'How often do you meet your target goals at work?': 'MeetTargets',
    'Does your emotional stability affect your workplace performance?': 'EmoStability',
    'In your opinion, how effective is team collaboration in promoting Emotional Intelligence?': 'TeamCollaboration'
}

# Rename the columns
df.rename(columns=column_mapping, inplace=True)

# Check the updated column names
print("\nUpdated column names:")
print(df.columns.tolist())
```

# **Data Visualization**

# 1) *Visualization of the Distribution of Age Groups*

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of age groups
plt.figure(figsize=(8, 6))
sns.countplot(x='Age', data=df, order=['18–25', '26–35', '36–45', '46 and above'])
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

# 2) *Visualization Emotional States at Work*

```python
# Split and stack the EmoStates column
emotional_states = df['EmoStates'].str.split(', ', expand=True).stack().value_counts()

# Visualize the most common emotional states
plt.figure(figsize=(10, 6))
emotional_states.plot(kind='bar')
plt.title('Most Common Emotional States at Work')
plt.xlabel('Emotional State')
plt.ylabel('Frequency')
plt.show()
```

# 3) *Visualization of Emotional Awareness vs. Productivity*

```python
# Visualize the relationship between emotional awareness and meeting targets
plt.figure(figsize=(10, 6))
sns.countplot(x='EmoAwareness', hue='MeetTargets', data=df)
plt.title('Emotional Awareness vs. Meeting Target Goals')
plt.xlabel('Emotional Awareness')
plt.ylabel('Count')
plt.legend(title='Meeting Targets')
plt.show()
```

# 4) *Visualization of Stress Management vs. Productivity*

```python
# Convert MeetTargets to numerical values for correlation analysis
df['MeetTargets_numeric'] = df['MeetTargets'].map({'Very often': 5, 'Often': 4, 'Sometimes': 3, 'Rarely': 2, 'Never': 1})

# Calculate correlation between stress management and meeting targets
correlation = df['StressManagement'].corr(df['MeetTargets_numeric'])
print(f"\nCorrelation between stress management and meeting targets: {correlation:.2f}")

# Visualize the relationship using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='StressManagement', y='MeetTargets_numeric', data=df)
plt.title('Stress Management vs. Meeting Targets')
plt.xlabel('Stress Management (1 = Poor, 5 = Excellent)')
plt.ylabel('Meeting Targets (1 = Never, 5 = Very often)')
plt.show()
```

# 5) *Visualization of Emotional Awareness Vs Productivity*

```python
# Stacked Bar Chart: EmoAwareness vs. MeetTargets
plt.figure(figsize=(12, 6))
pd.crosstab(df['EmoAwareness'], df['MeetTargets']).plot(kind='bar', stacked=True)
plt.title('Emotional Awareness vs. Meeting Targets')
plt.xlabel('Emotional Awareness Level')
plt.ylabel('Count')
plt.legend(title='Meeting Targets')
plt.show()
```

# 6) *Tolerance/Patience vs. Stress Management*

```python
# Map tolerance levels to numerical values
tolerance_mapping = {
    'Poor': 1,
    'Below average': 2,
    'Average': 3,
    'Good': 4,
    'Excellent': 5
}

df['TolerancePatience_numeric'] = df['TolerancePatience'].map(tolerance_mapping)

# Check for missing values after mapping
print("Missing values in TolerancePatience_numeric:")
print(df['TolerancePatience_numeric'].isnull().sum())


plt.figure(figsize=(10, 6))
sns.regplot(
    x='StressManagement',
    y='TolerancePatience_numeric',
    data=df,
    x_jitter=0.1,
    scatter_kws={'alpha': 0.5}  # Optional: Makes points semi-transparent
)
plt.title('Stress Management vs. Tolerance/Patience')
plt.xlabel('Stress Management (1 = Poor, 5 = Excellent)')
plt.ylabel('Tolerance/Patience (1 = Poor, 5 = Excellent)')
plt.yticks(
    ticks=[1, 2, 3, 4, 5],
    labels=['Poor', 'Below Average', 'Average', 'Good', 'Excellent']
)
plt.show()
```

# 7) *Team Collaboration Effectiveness*

```python
import matplotlib.pyplot as plt
import numpy as np

collab_counts = df['TeamCollaboration'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
x = np.arange(len(collab_counts))
y = collab_counts.values

ax.stem(x, y, linefmt="grey", markerfmt="o", basefmt=" ")
ax.set_xticks(x)
ax.set_xticklabels(collab_counts.index, rotation=45)

plt.title("Perceived Effectiveness of Team Collaboration in Promoting EQ")
plt.ylabel("Count")
plt.show()

```

# 8) *Educational Level Vs Emotional Impacts*

```python
# Heatmap: Education vs. EmoImpact
heatmap_data = pd.crosstab(df['Education'], df['EmoImpact'])
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='Blues')
plt.title('Education Level vs. Belief in Emotional Impact on Work')
plt.xlabel('Belief in Emotional Impact')
plt.ylabel('Education Level')
plt.show()
```

# 9) *Emotional Stability and Productivity*

```python
# Calculate mean target achievement for each Emotional Stability category
mean_targets = df.groupby('EmoStability')['MeetTargets_numeric'].mean().reset_index()

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='EmoStability', y='MeetTargets_numeric', data=mean_targets, marker='o', linestyle='-')

# Labels and title
plt.title('Emotional Stability vs. Meeting Targets')
plt.xlabel('Emotional Stability Affects Performance')
plt.ylabel('Average Meeting Targets Score (1=Never, 5=Very often)')

# Show the plot
plt.show()
```

# 10) *Policy Effectiveness vs. Productivity*

```python
# Bar Chart: PolicyEffectiveness vs. MeetTargets
plt.figure(figsize=(12, 6))
sns.countplot(x='PolicyEffectiveness', hue='MeetTargets', data=df)
plt.title('Policy Effectiveness vs. Meeting Targets')
plt.xlabel('Perceived Effectiveness of EQ Policies')
plt.ylabel('Count')
plt.legend(title='Meeting Targets')
plt.show()
```

```python
# Define mappings for Likert scale responses (modify as per your dataset)
#stress_mapping = {"Poor": 1, "Fair": 2, "Average": 3, "Good": 4, "Excellent": 5}
targets_mapping = {"Never": 1, "Rarely": 2, "Sometimes": 3, "Often": 4, "Very often": 5}

# Apply mappings
#df["StressManagement"] = df["StressManagement"].map(stress_mapping)
df["MeetTargets"] = df["MeetTargets"].map(targets_mapping)
```

```python
#df["StressManagement"] = pd.to_numeric(df["StressManagement"], errors="coerce")
df["MeetTargets"] = pd.to_numeric(df["MeetTargets"], errors="coerce")
```

```python
from scipy.stats import pearsonr

# Compute Pearson Correlation & p-value
correlation, p_value = pearsonr(df["StressManagement"], df["MeetTargets"])

print(f"Pearson Correlation: {correlation}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("✅ The correlation is statistically significant (Reject H₀)")
else:
    print("❌ The correlation is NOT statistically significant (Fail to Reject H₀)")
```
